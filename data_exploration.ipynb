```python
# filter out replies tweets
tweets = tweets[tweets["in_reply_to_tweetid"].isna()]
```

```python
tweets.shape
```

```python
# filter out quoted tweets
tweets = tweets[tweets["quoted_tweet_tweetid"].isna()]
```

```python
tweets["tweet_language"].value_counts()
```

```python
# filter all non-arabic tweets
tweets = tweets[tweets["tweet_language"] == "ar"]
```

```python
tweets = tweets.drop(
    columns=[
        "tweet_client_name",
        "in_reply_to_userid",
        "in_reply_to_tweetid",
        "quoted_tweet_tweetid",
        "retweet_userid",
        "retweet_tweetid",
        "latitude",
        "longitude",
        "tweet_language",
        "account_language",
        "tweet_time",
        "tweet_client_name",
    ]
)
```

```python
tweets.head()
```

```python
tweets.info()
```

```python
# filter all tweets that doesn't have any reaction info

tweets = tweets.dropna(
    subset=[
        "is_retweet",
        "quote_count",
        "reply_count",
        "reply_count",
        "like_count",
        "retweet_count",
    ]
)

# filter missing tweets

tweets = tweets.dropna(subset=["tweet_text"])
```
