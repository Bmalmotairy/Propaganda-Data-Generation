```python
sample_dfs = []
to_label_dfs = []
for cid in range(len(np.unique(clusters))):
    sample_df = (
        tweets[tweets["clusters"] == cid]
        .drop_duplicates(subset=["clean_tweet"])
        .sample(4_000, random_state=42)
    )
    sample_dfs.append(sample_df)
    to_label_dfs.append(sample_df.sample(150, random_state=42))
```

```python
propaganda = (
    pd.concat(sample_dfs, ignore_index=True)
    .drop(columns=["clusters"])
    .sample(frac=1.0)
    .reset_index(drop=True)
)
to_label = (
    pd.concat(to_label_dfs, ignore_index=True)
    .drop(columns=["clusters"])
    .sample(frac=1.0)
    .reset_index(drop=True)
)
```
