```python
import glob
```

```python
tweets_data_paths = glob.glob("./data/*tweets_csv*.csv")
```

```python
from tqdm.auto import tqdm

tweets_dfs = []

for tweet_data_path in tqdm(tweets_data_paths, total=len(tweets_data_paths)):
    print(f"Reading file {tweet_data_path}")
    tweets_dfs.append(pd.read_csv(tweet_data_path).drop_duplicates(["tweet_text"]))
```
